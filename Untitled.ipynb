# Topic: Trajectory of ball in Football

# Introduction

The trajectory of a football when struck with spin, especially the famous shot by Roberto Carlos during a 1997 match against France, demonstrates the Magnus effect, where the spin of the ball causes it to curve mid-flight. This particular shot, also known as a banana shot or curveball, curved unexpectedly around the defensive wall, confounding both the defenders and the goalkeeper.

Physics Behind the Curveball

When a football is struck with spin, the Magnus effect comes into play. As the ball spins, air pressure differences on either side of the ball cause it to deviate from a straight path. The side with faster airflow (due to spin) has lower pressure, while the opposite side has higher pressure, resulting in a curved trajectory.

# Key factors:

*Initial Velocity (v₀): The speed at which the ball is kicked.
    
*Spin (ω): The angular velocity applied to the ball.

*Magnus Effect: The lift force generated due to spin.

*Air Resistance: Opposing force depending on ball velocity and drag coefficient.

# Objective:

To mathematically model the path of Roberto Carlos’ curveball using differential equations and simulate its trajectory using Python.

# Differential Equation:

Let us denote the position of the ball at any time t by r(t) = (x(t), y(t), z(t)) , where:

    x(t)--> represents the horizontal position,
    y(t)--> represents the vertical position,
    z(t)--> represents the lateral (sideways) position.
}
The velocity vector is v(t) = $\frac{d\mathbf{r}(t)}{dt}$ \).

The differential equations governing the ball's motion are derived from Newton's second law:
\[
m $\frac{d^2 \mathbf{r}}{dt^2} = \mathbf{F}_\text{total}$,
\]
where \( m \) is the mass of the ball and \( $\mathbf{F}_\text{total}$ \) is the total force acting on the ball.

\subsection{Equations of Motion}
The total force includes gravitational force, drag force, and the Magnus force:
\[
$\mathbf{F}_\text{total} = \mathbf{F}_\text{gravity} + \mathbf{F}_\text{drag} + \mathbf{F}_\text{Magnus}$.
\]

The differential equations for each direction are:
\begin{align*}
m \frac{d^2 x}{dt^2} &= -C_d v_x \sqrt{v_x^2 + v_y^2 + v_z^2} + C_m \omega v_z, \\
m \frac{d^2 y}{dt^2} &= -m g - C_d v_y \sqrt{v_x^2 + v_y^2 + v_z^2}, \\
m \frac{d^2 z}{dt^2} &= -C_d v_z \sqrt{v_x^2 + v_y^2 + v_z^2} - C_m \omega v_x.
\end{align*}


\subsection{Magnus Effect}
The Magnus force is responsible for the ball's curved trajectory and is given by:
\[
$\mathbf{F}_\text{Magnus} = C_m (\mathbf{\omega} \times \mathbf{v}$),
\]
where $( C_m )$ is the Magnus coefficient, \( $\mathbf{\omega}$ \) is the angular velocity (spin), and \( $\mathbf{v}$ \) is the velocity of the ball.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Gravity (m/s^2)
C_d = 0.25  # Drag coefficient (estimated value)
C_m = 0.2   # Magnus coefficient (estimated value)
rho = 1.2   # Air density (kg/m^3)
A = 0.038   # Cross-sectional area of football (m^2)
m = 0.43    # Mass of football (kg)

# Differential equations for the system
def equations(y, t, omega):
    vx, vy, vz, x, y, z = y
    v = np.sqrt(vx**2 + vy**2 + vz**2)  # Velocity magnitude
    
    # Drag force components
    Fd_x = -0.5 * C_d * rho * A * v * vx / m
    Fd_y = -0.5 * C_d * rho * A * v * vy / m
    Fd_z = -0.5 * C_d * rho * A * v * vz / m
    
    # Magnus force components
    Fm_x = C_m * (omega[1] * vz - omega[2] * vy)
    Fm_y = C_m * (omega[2] * vx - omega[0] * vz)
    Fm_z = C_m * (omega[0] * vy - omega[1] * vx)
    
    # Equations of motion
    dvxdt = Fd_x + Fm_x
    dvydt = Fd_y + Fm_y - g
    dvzdt = Fd_z + Fm_z
    
    return [dvxdt, dvydt, dvzdt, vx, vy, vz]

# Initial conditions: [vx, vy, vz, x, y, z]
initial_conditions = [30, 0, 0, 0, 0, 0]  # Initial velocity and position
omega = np.array([0, 30, 0])  # Spin around the y-axis for curve

# Time
t = np.linspace(0, 5, 500)

# Solving the equations
solution = odeint(equations, initial_conditions, t, args=(omega,))

# Extracting position components
x = solution[:, 3]
y = solution[:, 4]
z = solution[:, 5]

# Plotting the trajectory
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('X Position (m)')
ax.set_ylabel('Y Position (m)')
ax.set_zlabel('Z Position (m)')
ax.set_title('Trajectory of Roberto Carlos Curveball')
plt.show()
```

*Spin (omega): The ball’s spin rate, which generates the Magnus force.


# Interpretation

The simulation shows that:

.Speed and Spin: The initial high velocity and rapid spin were responsible for the significant deflection of the ball. The ball curved inwards toward the goalpost after going wide.

.Magnus and Drag Forces: Both the Magnus and drag forces influenced the trajectory, with the Magnus effect dominating the curve and drag slowing down the ball over time.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_trajectory(initial_velocity, launch_angle, ball_diameter, ball_mass, coefficient_drag, time_step):
    """
    Calculates the trajectory of a soccer ball during a free kick.

    Args:
        initial_velocity: Initial velocity of the ball (m/s).
        launch_angle: Launch angle of the ball (degrees).
        ball_diameter: Diameter of the ball (m).
        ball_mass: Mass of the ball (kg).
        coefficient_drag: Coefficient of drag.
        time_step: Time step for calculations (s).

    Returns:
        A tuple containing the x and y positions of the ball over time.
    """

    # Convert launch angle to radians
    launch_angle = np.radians(launch_angle)

    # Calculate initial x and y velocities
    vx0 = initial_velocity * np.cos(launch_angle)
    vy0 = initial_velocity * np.sin(launch_angle)

    # Calculate cross-sectional area of the ball
    area = np.pi * (ball_diameter / 2)**2

    # Calculate air density (assuming standard conditions)
    air_density = 1.225

    # Initialize time, x, and y positions
    time = 0
    x = 0
    y = 0

    # Initialize x and y velocities
    vx = vx0
    vy = vy0

    # Initialize lists to store x and y positions over time
    x_positions = []
    y_positions = []

    # Calculate trajectory
    while y >= 0:
        # Calculate drag force
        drag_force = 0.5 * air_density * area * vx**2 * coefficient_drag

        # Calculate acceleration in x and y directions
        ax = -drag_force / ball_mass
        ay = -9.81 - drag_force / ball_mass

        # Update velocities
        vx += ax * time_step
        vy += ay * time_step

        # Update positions
        x += vx * time_step
        y += vy * time_step

        # Append positions to lists
        x_positions.append(x)
        y_positions.append(y)

        # Increment time
        time += time_step

    return x_positions, y_positions

# Define input parameters
initial_velocity = 30  # m/s
launch_angle = 45  # degrees
ball_diameter = 0.22  # m
ball_mass = 0.45  # kg
coefficient_drag = 0.45
time_step = 0.01  # s

# Calculate trajectory
x_positions, y_positions = calculate_trajectory(initial_velocity, launch_angle, ball_diameter, ball_mass, coefficient_drag, time_step)

# Plot trajectory  
plt.plot(x_positions, y_positions)
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.title("Soccer Ball Trajectory")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Gravity (m/s²)
air_density = 1.225  # Air density (kg/m³)
ball_mass = 0.43  # Mass of the ball (kg)
ball_diameter = 0.22  # Diameter of the ball (m)
ball_radius = ball_diameter / 2
area = np.pi * ball_radius**2  # Cross-sectional area of the ball (m²)
drag_coefficient = 0.25  # Coefficient of drag for a soccer ball
magnus_coefficient = 0.2  # Coefficient of Magnus effect (estimated)
time_step = 0.01  # Time step (s)

# Initial conditions
initial_velocity = 35  # Initial velocity of the ball (m/s)
launch_angle = 15  # Launch angle of the ball (degrees)
spin_rate = 15  # Spin rate of the ball (rad/s) for Magnus effect
vx0 = initial_velocity * np.cos(np.radians(launch_angle))  # Initial velocity in x-direction
vy0 = initial_velocity * np.sin(np.radians(launch_angle))  # Initial velocity in y-direction
vz0 = 0  # Initial velocity in z-direction (curve direction)

# Magnus force and drag force parameters
def magnus_force(vx, vz, spin_rate):
    """Calculates the Magnus force on the ball due to its spin."""
    magnus_magnitude = magnus_coefficient * spin_rate * ball_radius * np.sqrt(vx**2 + vz**2)
    magnus_fz = magnus_magnitude  # Lateral force causing the curve
    return magnus_fz

def drag_force(vx, vy):
    """Calculates the drag force on the ball."""
    v = np.sqrt(vx**2 + vy**2)
    drag_magnitude = 0.5 * drag_coefficient * air_density * area * v**2
    drag_fx = -drag_magnitude * vx / v
    drag_fy = -drag_magnitude * vy / v
    return drag_fx, drag_fy

# Lists to store trajectory
x_positions = []
y_positions = []
z_positions = []

# Initial conditions
x = 0  # Initial x position (m)
y = 6  # Initial y position (m) -> start from a height
z = 0  # Initial z position (sideways curve position)
vx = vx0  # Initial velocity in x-direction
vy = vy0  # Initial velocity in y-direction
vz = vz0  # Initial velocity in z-direction (sideways curve)

# Time loop to calculate the trajectory
while y >= 4:  # Stop when the ball reaches 4m height (around net height)
    # Calculate forces
    magnus_fz = magnus_force(vx, vz, spin_rate)
    drag_fx, drag_fy = drag_force(vx, vy)
    
    # Update velocities (with forces)
    ax = drag_fx / ball_mass
    ay = -g + drag_fy / ball_mass  # Include gravity in y-direction
    az = magnus_fz / ball_mass  # Magnus force causes the curve in z-direction

    # Update velocities
    vx += ax * time_step
    vy += ay * time_step
    vz += az * time_step

    # Update positions
    x += vx * time_step
    y += vy * time_step
    z += vz * time_step

    # Store the positions
    x_positions.append(x)
    y_positions.append(y)
    z_positions.append(z)

# Plotting the trajectory in 2D
plt.figure(figsize=(10, 5))

# Plot side view (x vs y)
plt.subplot(1, 2, 1)
plt.plot(x_positions, y_positions, label="Side View (x vs y)")
plt.xlabel("X Distance (m)")
plt.ylabel("Y Height (m)")
plt.title("Side View: Ball Trajectory (Banana Shot)")
plt.grid(True)
plt.legend()

# Plot top view (x vs z for curve)
plt.subplot(1, 2, 2)
plt.plot(x_positions, z_positions, label="Top View (x vs curve)", color="orange")
plt.xlabel("X Distance (m)")
plt.ylabel("Z Curve (m)")
plt.title("Top View: Ball Curving Trajectory")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
